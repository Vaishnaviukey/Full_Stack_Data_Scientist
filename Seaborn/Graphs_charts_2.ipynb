## Tips Analysis by using Graphs and Charts

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
import cufflinks as cnf
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
sns.relplot(x='total_bill',y='tip',data=df)
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='smoker') # hue gives the different color
```

```python
sns.relplot(x='total_bill',y='tip',data=df,style='size')
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='size')
```

```python
sns.relplot(x='total_bill',y='tip',data=df,style='smoker')#style provide different category or structure
```

```python
sns.relplot(x='total_bill',y='tip',data=df,style='time',hue='smoker')
```
